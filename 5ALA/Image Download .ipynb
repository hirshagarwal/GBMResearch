```python
import urllib.request, json
import requests
import os
```

```python
gene_name = "OCIAD2"
```

```python
# URLs
fetch_samples_url = "http://api.brain-map.org/api/v2/data/query.json?criteria=model::SectionDataSet,rma::criteria,treatments[name$eq\'ISH\'],genes[acronym$eq\'{0}\'],rma::include,genes,sub_images".format(gene_name)
download_ish_url = "http://api.brain-map.org/api/v2/image_download/{0}"
download_annotation_url = "http://api.brain-map.org/api/v2/image_download/{0}?view=tumor_feature_annotation"
download_expression_url = "http://api.brain-map.org/api/v2/image_download/{0}?view=expression"
```

```python
raw_json = ""
with urllib.request.urlopen(fetch_samples_url) as url:
    raw_json = json.loads(url.read().decode())
    print(raw_json)
```

```python
specimen_data = []
for specimen in raw_json["msg"]:
    data_tuple = (specimen["specimen_id"], specimen["sub_images"][0]["id"])
    specimen_data.append(data_tuple)
print("Downloading: {0} samples".format(len(specimen_data)))
for (specimen_id, image_id) in specimen_data:
    existing_files = os.listdir('images/auto{}'.format(gene_name))
    contains_ish = '{1}_ISH_{0}.jpg'.format(gene_name, specimen_id) in existing_files
    contains_annotation = '{1}_Annotation_{0}.jpg'.format(gene_name, specimen_id) in existing_files
    if (contains_ish & contains_annotation):
        print("Already contains: {0}".format(specimen_id))
        continue
    with open('images/auto{0}/{1}_ISH_{0}.jpg'.format(gene_name, specimen_id), 'wb') as f:
        image_url = download_ish_url.format(image_id)
        print(image_url)
        f.write(requests.get(image_url).content)
    with open('images/auto{0}/{1}_Annotation_{0}.jpg'.format(gene_name, specimen_id), 'wb') as f:
        image_url = download_annotation_url.format(image_id)
        print(image_url)
        f.write(requests.get(image_url).content)
#     with open('images/auto{0}/{1}_Expression_{0}.jpg'.format(gene_name, specimen_id), 'wb') as f:
#         image_url = download_expression_url.format(image_id)
#         print(image_url)
#         f.write(requests.get(image_url).content)
```
