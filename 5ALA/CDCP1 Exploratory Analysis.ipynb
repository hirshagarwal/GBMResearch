```python
from skimage import data, io, filters
from skimage.color import rgb2hed, hed2rgb
from PIL import Image
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.pyplot as plt
import numpy as np
Image.MAX_IMAGE_PIXELS = None
```

```python
ish_image = io.imread('images/CDCP1/W31_1_1_ISH_CDCP1.jpg')
```

```python
ish_split = rgb2hed(ish_image)
```

```python
cmap_hema = LinearSegmentedColormap.from_list('mycmap', ['white', 'navy'])
cmap_eosin = LinearSegmentedColormap.from_list('mycmap', ['darkviolet',
                                               'white'])
cmap_dab = LinearSegmentedColormap.from_list('mycmap', ['white',
                                             'saddlebrown'])
```

```python
fig, axes = plt.subplots(2, 2, figsize=(7, 6), sharex=True, sharey=True)
ax = axes.ravel()
ax[0].imshow(ish_image)
ax[1].imshow(ish_split[:, :, 0], cmap=cmap_hema)
ax[2].imshow(ish_split[:, :, 1], cmap=cmap_eosin)
ax[3].imshow(ish_split[:, :, 2], cmap=cmap_dab)
```
